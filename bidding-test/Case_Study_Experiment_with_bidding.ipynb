```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/vkoul/data/main/misc/ab_bidding.csv')

print(df.shape)
print(df.head())
print(df.info())
```

```python
df.groupby('Group')['Group'].value_counts()
```

```python
df.groupby('Group')[['Purchase', 'Earning']].mean()
```

```python
from scipy import stats

control = df[df['Group'] == 'control']['Purchase']
test = df[df['Group'] == 'test']['Purchase']

stat, p_value = stats.ttest_ind(control, test)
print(f'p-value: {p_value:.4f}')
```

```python

control_earning = df[df['Group'] == 'control']['Earning']
test_earning = df[df['Group'] == 'test']['Earning']

stat_e, p_value_e = stats.ttest_ind(control_earning, test_earning)
print(f'Earning — p-value: {p_value_e:.4f}')
print(f'Control mean Earning: {control_earning.mean():.2f}')
print(f'Test mean Earning: {test_earning.mean():.2f}')

lift = (test_earning.mean() - control_earning.mean()) / control_earning.mean() * 100
print(f'Lift: {lift:.1f}%')
```

The difference in purchase count between average bidding and maximum bidding is not statistically significant (p=0.35) — so on its own, that metric doesn't tell us much.
But purchase count isn't the metric that matters here — revenue is. Checking Earning, maximum bidding (test) generates 31.8% more revenue than average bidding (control), and this difference is highly significant (p≈0.0000).
Recommendation: Switch to maximum bidding. It doesn't produce more purchases, but each purchase is worth significantly more — the strategy is winning higher-value opportunities, not just more of them. Revenue, not purchase count, is the right metric to make this call on.
